# 6.Dictionary in Python

 * __Dictioanry are used to store data value in key:value pairs.__
 * __A dictionary is a collection which is ordered, changeable or mutable and do not allow duplicate.__
 * __Dictionary can not have two items with the same key.__
 * __A dictionary can be nested and contains another dictionary.__
 * __A dictionary items are presented in key:value pairs, and can be referred to by using key name.__
   
      * __"Keys":__-------__"Values"__
      * __"name":__-------__"xyz"__
      * __"age":__--------__25__
      * __"email":__------__"xyz25@hotmail.com"__
      * __"address":__----__"Pune"__
      * __"Contact":__----__"78945612310"
      * __"Job":__--------__"Frontend Developer"__

```python
# Take a list as input from the user.
# If the first and last element of the list is the same return as True
#If the first and last element of the list is not true return it as False. 


list = []
n = int(input("Enter number of elements :"))
for i in range (0,n):
    element = int(input())
    list.append(element)
print(list)

def list_check(list):
    if list[0]==list[-1]:
        return True
    else:
        return False
z = list_check(list)
print("result is",z)
```
